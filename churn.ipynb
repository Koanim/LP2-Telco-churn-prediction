```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.preprocessing import LabelEncoder



from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split


X_train, X_eval, y_train, y_eval = train_test_split(x, y, test_size=0.3, random_state=42,stratify=y)

encoder = LabelEncoder()
y_train_enc= encoder.fit_transform(y_train)
y_eval_enc= encoder.transform(y_eval)
# Categorical pipeline
categorical_pipeline = Pipeline(steps=[
    ('cat_imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Numerical pipeline
numeric_pipeline = Pipeline(steps=[
    ('num_imputer', SimpleImputer(strategy='mean')),
    ('scaler', RobustScaler())
])

# Create column transformer
preprocessor = ColumnTransformer(transformers=[
    ('numeric_pipeline_name', numeric_pipeline, list_num),  # Apply numeric pipeline to numerical columns
    ('categorical_pipeline_name', categorical_pipeline, list_cat)  # Apply categorical pipeline to categorical columns
])


```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Define the models
models = {}
models['Logistic_Regression'] = LogisticRegression(random_state=42)
#models['Support_Vector_Machine'] = LinearSVC(random_state=42)
models['Decision_Tree'] = DecisionTreeClassifier(random_state=42)
models['Random_Forest'] = RandomForestClassifier(random_state=42)
models['Naive_Bayes'] = GaussianNB()
models['K-Nearest_Neighbors'] = KNeighborsClassifier()
models['Gradient_Boosting'] = GradientBoostingClassifier(random_state=42)

```

```python
full_pipelines = {}
for name, model in models.items():
    full_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),  # Preprocessing
        ('model', model)  # Model
    ])
    full_pipelines[name] = full_pipeline

# Fit the models and store their accuracies
accuracies = {}
for name, pipeline in full_pipelines.items():
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_eval)
    accuracies[name] = accuracy_score(y_eval, y_pred)

# Print the accuracies
for name, acc in accuracies.items():
    print(f"{name}: {acc}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
import pandas as pd

# Assuming X_train, y_train_encode, X_eval, and y_eval_encode are defined

all_pipelines = {}
metric_table = pd.DataFrame(columns=['model_name', 'accuracy', 'precision', 'recall', 'f1-score'])

for name, model in models.items():
    final_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])

    final_pipeline.fit(X_train, y_train_enc)
    all_pipelines[name] = final_pipeline

    y_pred = final_pipeline.predict(X_eval)

    metrics = classification_report(y_eval_enc, y_pred, output_dict=True)

    accuracy = metrics['accuracy']
    precision = metrics['weighted avg']['precision']
    recall = metrics['weighted avg']['recall']
    f1_score = metrics['weighted avg']['f1-score']

    metric_table.loc[len(metric_table)] = [model, accuracy, precision, recall, f1_score]

# Sort the metric_table based on f1-score in descending order
metric_table = metric_table.sort_values(by='f1-score', ascending=False)

print(metric_table)
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as iPipeline
from sklearn.feature_selection import mutual_info_classif,SelectKBest
from sklearn.metrics import confusion_matrix


import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

# Define models
models = {}
models['Logistic_Regression'] = LogisticRegression(random_state=42)
#models['Support_Vector_Machine'] = LinearSVC(random_state=42)
models['Decision_Tree'] = DecisionTreeClassifier(random_state=42)
models['Random_Forest'] = RandomForestClassifier(random_state=42)
models['Naive_Bayes'] = GaussianNB()
models['K-Nearest_Neighbors'] = KNeighborsClassifier()
models['Gradient_Boosting'] = GradientBoostingClassifier(random_state=42)



all_balanced_pip = {}
all_confusion_matrix = {}

balanced_metric_table = pd.DataFrame(columns=['model_name', 'accuracy', 'precision', 'recall', 'f1_score'])

for name, model in models.items():

    balanced_pipeline = iPipeline(steps=[
        ('preprocessor', preprocessor),
        ('smote_sampler', SMOTE(random_state=0)),
        ('feature_selection', SelectKBest(mutual_info_classif, k='all')),
        ('model', model)
    ])

    balanced_pipeline.fit(X_train, y_train_enc)
    all_balanced_pip[name] = balanced_pipeline

    y_pred = balanced_pipeline.predict(X_eval)

    conf_matrix = confusion_matrix(y_eval_enc, y_pred)
    all_confusion_matrix[name] = conf_matrix

    balanced_metrics = classification_report(y_eval_enc, y_pred, output_dict=True)

    accuracy = balanced_metrics['accuracy']
    precision = balanced_metrics['weighted avg']['precision']
    recall = balanced_metrics['weighted avg']['recall']
    f1_score = balanced_metrics['weighted avg']['f1-score']

    balanced_metric_table.loc[len(balanced_metric_table)] = [name,accuracy,precision,recall,f1_score]

    # Plot confusion matrix for Random Forest
    if name == 'Random_Forest':  # Fixing the condition
        sns.heatmap(conf_matrix, annot=True, cmap='coolwarm', fmt='.1f')
        plt.show()

    if name == 'LogisticRegression' :
         sns.heatmap(conf_matrix, annot=True, cmap='coolwarm', fmt='.1f')
         plt.show()

    if name == 'Gradient_Boosting' :
         sns.heatmap(conf_matrix, annot=True, cmap='coolwarm', fmt='.1f')
         plt.show()
    


```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import roc_curve, auc

# Create a figure and axis for plotting
fig, ax = plt.subplots(figsize=(8, 8))

# Dictionary to store ROC data for each model
roc_data = {}

# Iterate over each model in all_balanced_pipelines
for name, pipeline in all_balanced_pip.items():
    # Get predicted probabilities for the positive class
    y_score = pipeline.predict_proba(X_eval)[:, 1]
    
    # Compute ROC curve and AUC
    fpr, tpr, thresholds = roc_curve(y_eval_enc, y_score)
    roc_auc = auc(fpr, tpr)
    
    # Create a DataFrame to store ROC data
    roc_data_df = pd.DataFrame({'False Positive Rate': fpr, 'True Positive Rate': tpr, 'Threshold': thresholds})
    roc_data[name] = roc_data_df
    
    # Plot ROC curve
    ax.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

# Plot the diagonal line (random classifier)
ax.plot([0, 1], [0, 1], linestyle='--', color='gray')

# Set labels and title
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('Receiver Operating Characteristic (ROC) Curve')
ax.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
from scipy import stats

# Assuming df is your DataFrame with columns 'StreamingTV' and 'Churn'
# Split the data into two groups based on 'StreamingTV'
group_yes = result[result['StreamingTV'] == 'Yes']['Churn']
group_no = result[result['StreamingTV'] == 'No']['Churn']

# Calculate the mean for each group
mean_yes = group_yes.mean()
mean_no = group_no.mean()

# Perform a t-test assuming unequal variances
t_stat, p_value = stats.ttest_ind(group_yes, group_no, alternative='less')

# Define the significance level (commonly 0.05)
alpha = 0.05

# Print the means and t-test results
print("Mean Churn (StreamingTV = 'Yes'):", mean_yes)
print("Mean Churn (StreamingTV = 'No'):", mean_no)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence that the average churn for 'StreamingTV' = 'Yes' is less than 'No'.")
else:
    print("Fail to reject the null hypothesis. There is no evidence that the average churn for 'StreamingTV' = 'Yes' is less than 'No'.")


```
